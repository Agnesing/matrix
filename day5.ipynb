```
def run_model (feats, model = DecisionTreeRegressor(max_depth=5)):
  X = df[feats].values
  y = df['prices_amountmin'].values

  scores = cross_val_score (model, X, y, scoring = 'neg_mean_absolute_error')
  return np.mean(scores), np.std(scores)
```

Kolumna features zawiera słowniki zapisane w postacji string

mając słownik i podając metodę str(dict) otrzymamy zapis stringowy słownika, analogiczny jak w kolumnie features. 

Poniższa funkcja przywraca postać słownika ze stringa

```
str_dict = '[{"key":"Gender","value":["Men"]},{"key":"Shoe Size","value":["M"]},{"key":"Shoe Category","value":["Men\'s Shoes"]},{"key":"Color","value":["Multicolor"]},{"key":"Manufacturer Part Number","value":["8190-W-NAVY-7.5"]},{"key":"Brand","value":["Josmo"]}]'
literal_eval(str_dict)
```

```
def parse_features (x):
  output_dict = {}
  if str(x) == 'nan' : return output_dict
  features = literal_eval(x.replace('\\"','"'))
  #po tej operacji mamy 'key': 'Gender', 'value': ['Men'], a chcemy 'Gender': 'Men'
  
  for item in features:
    key = item['key'].lower().strip() #to co jest pod key przypisujemy do key 
    # i jednoczesnie zamieniamy na małe litery i usuwamy spacje
    value = item['value'][0].lower().strip()
    
    output_dict[key]=value #wynik zapisujemy do nowego słownika

  return output_dict

df['features_parsed']= df['features'].map(parse_features)
```

```
df['features_parsed'].head()
```

```
keys = set()
df['features_parsed'].map (lambda x:keys.update(x.keys()))
len(keys)
```

W kolejnym kroku dodajemy kolumny do danych odpowiadające poszczególnym key w zbiorze keys. Dostaną one nową nazwę z przedrostkiem "feat_".

```
def get_name_feat(key):
  return 'feat_' + key

for key in tqdm_notebook(keys):
  df[get_name_feat(key)]=df.features_parsed.map(lambda feats: feats[key] if key in feats else np.nan)
```

```
df.columns
```

Następnie sprawdzamy jak często występują poszczególne cechy.

```
keys_stat = {}
for key in keys:
  keys_stat[key] = df [False == df[get_name_feat(key)].isnull()].shape[0]/df.shape[0]*100
```

Danych jest dużo. Tworzymy filtr, który wybiera tylko te czechy które występują w ponad 30% danych.

```
{k:v for k,v in keys_stat.items() if v>30}
```
